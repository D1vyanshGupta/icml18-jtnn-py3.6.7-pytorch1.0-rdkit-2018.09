```python
import os, pickle
import numpy as np
import matplotlib.pyplot as plt

LOSS_ORG_PATH = '/Users/divyanshgupta/QM9/logs/jtvae_qm9_loss_lst.pkl'
LOSS_GC_PATH = '/Users/divyanshgupta/QM9/logs/jtvae_+_graph_conv_+_tree_conv_qm9_loss_lst.pkl'

WORD_ORG_PATH = '/Users/divyanshgupta/QM9/logs/jtvae_qm9_wacc_lst.pkl'
WORD_GC_PATH = '/Users/divyanshgupta/QM9/logs/jtvae_+_graph_conv_+_tree_conv_qm9_wacc_lst.pkl'

TOPO_ORG_PATH = '/Users/divyanshgupta/QM9/logs/jtvae_qm9_tacc_lst.pkl'
TOPO_GC_PATH = '/Users/divyanshgupta/QM9/logs/jtvae_+_graph_conv_+_tree_conv_qm9_tacc_lst.pkl'

ASSM_ORG_PATH = '/Users/divyanshgupta/QM9/logs/jtvae_qm9_aacc_lst.pkl'
ASSM_GC_PATH = '/Users/divyanshgupta/QM9/logs/jtvae_+_graph_conv_+_tree_conv_qm9_aacc_lst.pkl'

KL_ORG_PATH = '/Users/divyanshgupta/QM9/logs/jtvae_qm9_kl_div_lst.pkl'
KL_GC_PATH = '/Users/divyanshgupta/QM9/logs/jtvae_+_graph_conv_+_tree_conv_qm9_kl_div_lst.pkl'

PARENT_DIR_PATH = os.getcwd()
```

```python
org_loss = []
with open(LOSS_ORG_PATH, 'rb') as file:
    org_loss = pickle.load(file)
    
gc_loss = []
with open(LOSS_GC_PATH, 'rb') as file:
    gc_loss = pickle.load(file)

fig = plt.figure(0, figsize=(15, 10))
plt.plot(np.arange(len(org_loss)), org_loss, label='Overall Loss for original JT-VAE')
plt.plot(np.arange(len(gc_loss)), gc_loss, label='Overall Loss for Graph ConvNet JT-VAE')

plt.title("Overall Loss for both models", fontsize=20, pad=15.0)

plt.xlabel('Iterations', fontsize=20, labelpad=10)
plt.ylabel('Loss', fontsize=20, labelpad=10)

plt.tick_params(labelsize=20)
plt.legend()

plot_name = "loss_comparison"

fig_path = os.path.join(PARENT_DIR_PATH, 'plots', plot_name + '.png')
plt.savefig(fig_path, dpi=200)
plt.close(fig)
```
